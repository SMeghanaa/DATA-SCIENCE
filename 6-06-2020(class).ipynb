```
import pandas as pd
import numpy as np
```

```
s1=pd.Series([9,5,6,7,8,7,3,2,0])
s2=pd.Series([1.2,3.4,5.6,8.9,9.0,1.2,4.3,6.5])
s3=pd.Series(['a','b','c','d','e','f','g','h','i'])
data={'first':s1,'second':s2,'third':s3}
df=pd.DataFrame(data)
df
```

```
df.rename(columns={"second":"column2","third":"column3"},inplace=True)
```

```
df
```

####Changing the row index:
1. set_index()
2. reset_index()

```
df=df.set_index('column3')
df
```

```
df=df.reset_index()
df     
```

```
#making alphabets as the index
label=list("abcdefghi")
label
```

```
df=df.set_index([pd.Index(label)])
df
```

####Removing columns 
1. drop()

```
#single column
df.drop(columns=["column3"])             #axis=1 for column and axis=0 for rows
df                                                 #inplace=True,removes the column permenantly
```

```
df.drop(columns=["first","column2"])
```

```
df.drop(df.columns[[0,2]],axis=1)    #column position,0=1st column,2=3rd column
```
