```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas.util.testing as tm
```

```
from sklearn.datasets import load_boston
boston=load_boston()
```

```
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
boston_df.head()
```

```
boston_df['MEDV']=boston.target
boston_df.head()
```

Split the data into training and test with a split ratio of 80-20(80% of the data into training data and 20% of the data into testing)

```
## split the data into input  nd output
x=boston_df.drop(['MEDV'],axis=1)
y=boston_df.MEDV
```

```
## divide the x and y into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.80,random_state=0)
```

```
## build the model on training data--->x_train,y_train
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```
x_train
```

```
y_train_pred=lm.predict(x_train)
y_train_pred
```

Apply feature scaling on the input columns, and build the Multiple Linear Regression Model. Check the performance on training and testing. Write your inferences on performance metrics and error metrics.

```
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("r2_score :",r2_score(y_train,y_train_pred))
print("Adjusted r2_score",1-((1-r2_score(y_train,y_train_pred))*(len(x_train)-1))/((len(x_train)-x_train.shape[1]-1)))
print("difference between r2_score and Adjusted r2_score",r2_score(y_train,y_train_pred)-(1-((1-r2_score(y_train,y_train_pred))*(len(x_train)-1))/((len(x_train)-x_train.shape[1]-1))))
len(x_train)==x_train.shape[0]#rows
print("MAE:",mean_absolute_error(y_train,y_train_pred))
print("MSE:",mean_squared_error(y_train,y_train_pred))
print("RMSE",np.sqrt(mean_squared_error(y_train,y_train_pred)))
```

```
# test data-unseen data
y_test_pred = lm.predict(x_test)
y_test_pred
```

```
## r2_value -->
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
print("r2_score :",r2_score(y_test,y_test_pred))
print("Adjusted r2_score",1-((1-r2_score(y_test,y_test_pred))*(len(x_test)-1))/((len(x_test)-x_test.shape[1]-1)))
print("difference between r2_score and Adjusted r2_score",r2_score(y_test,y_test_pred)-(1-((1-r2_score(y_test,y_test_pred))*(len(x_test)-1))/((len(x_test)-x_test.shape[1]-1))))
print("MAE:",mean_absolute_error(y_test,y_test_pred))
print("MSE:",mean_squared_error(y_test,y_test_pred))
print("RMSE",np.sqrt(mean_squared_error(y_test,y_test_pred)))
```
